```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = {
    "Fecha": pd.date_range(start="2025-01-01", periods=10, freq="D"),
    "Producto": ["Helado Chocolate", "Helado Vainilla", "Helaso Fresa",
                 "Helado Chocolate", "Paletas", "Helado Oreo",
                 "Brownie", "Helado Chocolate", "Cheesecake", "Paletas"],
    "Precio":  [180, 150, 160, 180, 120, 200, 220, 180, 250, 120],
    "Ventas": [30, 25, 20, 35, 40, 28, 15, 45, 10, 50]
}

df = pd.DataFrame(datos)
df
```

```python
df["Ingreso"] = df["Precio"] * df["Ventas"]
df
```

```python
ventas_diarias = df.groupby("Fecha")["Ingreso"].sum()
ventas_diarias
```

```python
plt.plot(ventas_diarias.index, ventas_diarias.values)
plt.title("Ingresos diarios")
plt.xlabel("Fecha")
plt.ylabel("Ingreso")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Mes"] = df["Fecha"].dt.month
resumen_mensual = df.groupby("Mes")["Ingreso"].sum()
resumen_mensual
```
