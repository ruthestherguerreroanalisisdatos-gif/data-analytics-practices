```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = {
    "Producto": ["Helado Vainilla", "Helado Chocolate", "Helado Fresa",
                 "Helado Oreo", "Paletas", "Brownie", "Cheesecake"],
    "Categoria": ["Helados", "Helados", "Helados",
                  "Helados", "Paletas", "Postres", "Postres"],
    "Precio": [150, 180, 160, 200, 120, 220, 250],
    "Ventas": [120, 180, 90, 140, 220, 80, 60]
}

df = pd.DataFrame(datos)
df
```

```python
df["Ingreso"] = df["Precio"] * df["Ventas"]
df
```

```python
resumen_categoria = df.groupby("Categoria").agg(
    Ventas_totales=("Ventas", "sum"),
    Ingreso_total=("Ingreso", "sum")
)

resumen_categoria
```

```python
plt.bar(resumen_categoria.index, resumen_categoria["Ingreso_total"])
plt.title("Ingreso total por categoria")
plt.show()
```

## Qué categoría genera más ingresos
Segun las proyecciones de ventas la categoría más demanda es el helado, quienes dominan nuestro mercado de ventas.
