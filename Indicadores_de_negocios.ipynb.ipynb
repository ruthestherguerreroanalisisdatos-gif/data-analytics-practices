```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = {
    "Clientes":  ["Ana", "Luis", "María", "Carlos", "Ana",
                  "Luis", "Pedro", "María", "Carlos", "Ana"],
    "Producto":  ["Helado Chocolate", "Helado Vainilla", "Paletas",
                  "Helado Oreo", "Brownie", "Cheesecake",
                  "Helado Chocolate", "Helado Fresa", "Paletas", "Helado Oreo"],
"Precio":  [180, 150, 120, 200, 220, 250, 180, 160, 120, 200],
    "Cantidad":  [2, 3, 4, 1, 2,1, 3, 2, 5, 1]
}

df = pd.DataFrame(datos)
df
```

```python
df ["Ingreso"] = df ["Precio"] * df ["Cantidad"]
df
```

```python
ingreso_total = df ["Ingreso"].sum()
ingreso_total
```

```python
ticket_promedio = df ["Ingreso"].mean()
ticket_promedio
```

```python
ventas_clientes = df.groupby("Clientes")["Ingreso"].sum()
ventas_clientes
```

```python
ventas_clientes.sort_values(ascending=False)
```

```python
ventas_clientes.plot(kind="bar", title="Ingreso por clientes")
plt.show()
```
