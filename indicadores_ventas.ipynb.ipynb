```python
import pandas as pd

datos = {
    "Producto": ["Vainilla", "Chocolate", "Fresa", "Mango", "Oreo"],
    "Precio": [150, 180, 160, 170, 200],
    "Ventas": [20, 35, 25, 30, 15]
}

df = pd.DataFrame(datos)
df
```

```python
df["Ingreso"] = df["Precio"] * df["Ventas"]
df
```

```python
ingreso_total = df["Ingreso"].sum()
ingreso_total
```

```python
precio_promedio = df["Precio"].mean()
precio_promedio
```

```python
producto_estrella = df.loc[df["Ingreso"].idxmax(), "Precio"]
producto_estrella
```

```python
ventas_totales = df["Ventas"].sum()
ventas_totales
```

```python
kpis = pd.DataFrame({
    "Indicador": [
        "Ingreso total",
        "Ventas totales",
        "precio promedio",
        "producto estrella"
    ],
    "Valor": [
        ingreso_total,
        ventas_totales,
        round(precio_promedio, 2),
        producto_estrella
    ]
})

kpis
```

```python
import matplotlib.pyplot as plt
df.plot(x="Producto", y="Ingreso", kind="bar")
plt.title("Ingreso por producto")
plt.ylabel("Ingreso")
plt.show()
```

## Reporte Ejecutivo KPIs
 El ingreso total del período fue **${ingreso_total}**.
- El producto estrella fue **{producto_estrella}**.
- El precio promedio de los productos fue **${precio_promedio:.2f}**.

Estos KPIs permiten evaluar el desempeño del negocio y apoyar la toma de decisiones.
