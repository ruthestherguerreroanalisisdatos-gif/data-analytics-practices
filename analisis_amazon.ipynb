```python
import pandas as pd

# Precios
df["discounted_price"] = (
    df["discounted_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors="coerce")

df["actual_price"] = (
    df["actual_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)
df["actual_price"] = pd.to_numeric(df["actual_price"], errors="coerce")

# Descuento
df["discount_percentage"] = (
    df["discount_percentage"]
    .astype(str)
    .str.replace("%", "", regex=False)
)
df["discount_percentage"] = pd.to_numeric(df["discount_percentage"], errors="coerce")

# Rating (limpio)
df["rating_clean"] = (
    df["rating"]
    .astype(str)
    .str.extract(r"(\d\.\d)")
)
df["rating_clean"] = pd.to_numeric(df["rating_clean"], errors="coerce")

# Rating count
df["rating_count"] = (
    df["rating_count"]
    .astype(str)
    .str.replace(",", "", regex=False)
)
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")
```

```python
df.info()
```

```python
top_categories = (
    df.groupby("category")["rating_count"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_categories
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16, 10))

# KPIs
ax_kpi = plt.subplot2grid((2, 3), (0, 0), colspan=3)
ax_kpi.axis("off")

ax_kpi.text(
    0.5, 0.5,
    f"Productos\n{df['product_id'].nunique()}\n\n"
    f"Precio promedio\n{df['discounted_price'].mean():.2f}\n\n"
    f"Descuento promedio (%)\n{df['discount_percentage'].mean():.2f}\n\n"
    f"Rating promedio\n{df['rating_clean'].mean():.2f}",
    ha="center", va="center", fontsize=14
)

# Gráfico categorías
ax1 = plt.subplot2grid((2, 3), (1, 0))
ax1.bar(top_categories.index, top_categories.values)
ax1.set_title("Top 5 Categorías por Reseñas")
ax1.set_ylabel("Reseñas")
ax1.tick_params(axis="x", rotation=45)

# Precio vs rating
ax2 = plt.subplot2grid((2, 3), (1, 1))
ax2.scatter(df["discounted_price"], df["rating_clean"])
ax2.set_title("Precio vs Rating")

# Top productos
top_products = (
    df.sort_values("rating_clean", ascending=False)
    .head(10)
)

ax3 = plt.subplot2grid((2, 3), (1, 2))
ax3.barh(top_products["product_name"], top_products["rating_clean"])
ax3.invert_yaxis()
ax3.set_title("Top 10 Productos")

plt.tight_layout()
plt.show()
```

```python
# KPIs
total_products = df.shape[0]
avg_price = df["discounted_price"].mean()
avg_discount = df["discount_percentage"].mean()
avg_rating = df["rating_clean"].mean()
```

```python
import matplotlib.pyplot as plt

# Estilo limpio
plt.style.use("seaborn-v0_8-whitegrid")

# Figura principal
fig = plt.figure(figsize=(20, 12))

# ---------------- KPIs ----------------
ax_kpi = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax_kpi.axis("off")

kpi_text = (
    f"Productos analizados: {total_products:,}\n"
    f"Precio promedio: ₹{avg_price:,.2f}\n"
    f"Descuento promedio: {avg_discount:.1f}%\n"
    f"Rating promedio: {avg_rating:.2f}"
)

ax_kpi.text(
    0.5, 0.5,
    kpi_text,
    ha="center", va="center",
    fontsize=18,
    bbox=dict(boxstyle="round,pad=0.6", facecolor="#F5F5F5")
)

# ---------------- Gráfico 1 ----------------
ax1 = plt.subplot2grid((3, 3), (1, 0))
top_categories.plot(kind="bar", ax=ax1)
ax1.set_title("Top 5 Categorías por Reseñas", fontsize=14)
ax1.set_ylabel("Cantidad de reseñas")
ax1.tick_params(axis='x', rotation=45)

# ---------------- Gráfico 2 ----------------
ax2 = plt.subplot2grid((3, 3), (1, 1))
df["rating_clean"].plot(kind="hist", bins=20, ax=ax2)
ax2.set_title("Distribución de Ratings", fontsize=14)
ax2.set_xlabel("Rating")

# ---------------- Gráfico 3 ----------------
ax3 = plt.subplot2grid((3, 3), (1, 2))
top_products.plot(kind="barh", ax=ax3)
ax3.set_title("Top 10 Productos más valorados", fontsize=14)
ax3.invert_yaxis()

# ---------------- Gráfico 4 ----------------
ax4 = plt.subplot2grid((3, 3), (2, 0), colspan=3)
df.groupby("category")["discount_percentage"].mean().sort_values(ascending=False).head(10).plot(
    kind="bar", ax=ax4
)
ax4.set_title("Top 10 Categorías con Mayor Descuento Promedio", fontsize=14)
ax4.set_ylabel("Descuento (%)")
ax4.tick_params(axis='x', rotation=45)

# Ajuste final
plt.subplots_adjust(hspace=0.5, wspace=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt

def kpi_card(ax, title, value, subtitle="", bg="#F8F9FA"):
    ax.axis("off")
    
    # Caja
    ax.text(
        0.5, 0.6, value,
        ha="center", va="center",
        fontsize=24, fontweight="bold",
        bbox=dict(boxstyle="round,pad=0.6", facecolor=bg, edgecolor="#DADADA")
    )
    
    # Título
    ax.text(
        0.5, 0.85, title,
        ha="center", va="center",
        fontsize=12, color="#555555"
    )
    
    # Subtítulo
    if subtitle:
        ax.text(
            0.5, 0.3, subtitle,
            ha="center", va="center",
            fontsize=10, color="#777777"
        )
```
