```python
import pandas as pd
df = pd.read_csv(r"C:\Users\lapto\OneDrive\Documents\ventas_heladeria.csv")
df["Ingreso"] = df["Precio"] * df["Ventas"]
df
```

```python
df.groupby("Producto")["Ingreso"].sum()
```

```python
df.groupby("Producto").agg({
    "Ventas": ["sum", "mean"],
    "Ingreso": ["sum", "mean"]
})
```

```python
resumen = df.groupby("Producto")["Ingreso"].sum().sort_values(ascending=False)
resumen.to_csv("resumen_ingreso.csv")
resumen
```
