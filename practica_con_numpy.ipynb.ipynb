```python
import pandas as pd
import numpy as np
```

```python
datos = {
    "Producto": ["Helado Chocolate", "Helado Vainilla", None,
                 "Helado Oreo", "Paletas", "Helado Chocolate"],
    "Precio": [180, 150, 160, None, 120, -180],
    "Ventas": [30, None, 20, 25, -5, 40]
}

df = pd.DataFrame(datos)
df
    
```

```python
df.isnull().sum()
```

```python
df["Producto"] = df["Producto"].fillna("Desconocido")
df["Precio"] = df["Precio"].fillna(df["Precio"].mean())
df["Ventas"] = df["Ventas"].fillna(0)

df
```

```python
df["Precio"] = df["Precio"].abs()
df["Ventas"] = df["Ventas"].abs()

df
```

```python
df["Ingreso"] = df["Precio"] * df["Ventas"]
df
```
