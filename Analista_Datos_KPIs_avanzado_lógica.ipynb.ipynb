```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = {
    "Fecha": pd.date_range(start="2025-01-01", periods=15, freq="D"),
    "Producto": [
        "Helado Chocolate","Helado Vainilla","Helado Fresa",
        "Helado Chocolate","Paletas","Helado Oreo",
        "Brownie","Helado Chocolate","Cheesecake","Paletas",
        "Helado Oreo","Brownie","Helado Vainilla","Paletas","Cheesecake"
    ],
    "Categoria": [
        "Helados","Helados","Helados",
        "Helados","Paletas","Helados",
        "Postres","Helados","Postres","Paletas",
        "Helados","Postres","Helados","Paletas","Postres"
    ],
    "Precio": [180,150,160,180,120,200,220,180,250,120,200,220,150,120,250],
    "Cantidad": [3,2,2,4,5,3,1,6,1,7,2,2,3,6,2]
}

df = pd.DataFrame(datos)
df
```

```python
df["Ingreso"] = df["Precio"] * df["Cantidad"]
df
```

```python
ticket_diario = df.groupby("Fecha")["Ingreso"].mean()
ticket_diario
```

```python
ingreso_categoria = df.groupby("Categoria")["Ingreso"].sum()
ingreso_categoria
```

```python
df.groupby("Producto")["Ingreso"].sum().sort_values(ascending=False)
```

```python
ingreso_categoria.plot(kind="bar", title="Ingreso por categoria")
plt.show()
```
