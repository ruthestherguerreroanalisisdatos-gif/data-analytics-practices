```python
import pandas as pd
Clientes = {
    "Clientes": ["Ana", "Luis", "María", "Pedro", "Juan", "Sofía", "Carlos", "Lucía"],
    "Frecuencia_compras": [12, 5, 20, 3, 15, 8, 2, 18],
    "Gasto_promedio": [350, 220, 500, 180, 420, 300, 150,480]
}

df_Clientes = pd.DataFrame(Clientes)
df_Clientes
```

```python
df_Clientes["Valor_Cliente"] = (
    df_Clientes["Frecuencia_compras"] * df_Clientes["Gasto_promedio"]
)

df_Clientes
```

```python
df_Clientes.sort_values(by="Valor_Cliente", ascending=False)
```

```python
def clasificar_cliente(valor):
    if valor >= 8000:
        return "Cliente Premium"
    elif valor >= 4000:
        return "Cliente Frecuente"
    else:
        return "Cliente Ocasional"

df_Clientes["Segmento"] = df_Clientes["Valor_Cliente"].apply(clasificar_cliente)
df_Clientes
```

```python
df_Clientes.groupby("Segmento")["Valor_Cliente"].mean()
```

Los clientes Premium generan el mayor valor para la empresa, por lo que se recomienda priorizar estrategias de fidelización y beneficios exclusivos. Los clientes ocasionales representan una oportunidad de crecimiento mediante promociones y campañas dirigidas.
