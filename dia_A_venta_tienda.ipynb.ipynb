```python
import pandas as pd

datos = {
    "Fecha": ["2025-01-01", "2025-01-01", "2025-01-02", "2025-01-02", "2025-01-03"],
    "Producto": ["Camisa", "Zapatos", "Camisa", "Pantalón", "Zapatos"],
    "Precio": [25, 60, 25, 40, 60],
    "Cantidad": [4, 2, 3, 1, 5]
}

df = pd.DataFrame(datos)
df
```

```python
df["Ingreso"] = df["Precio"] * df["Cantidad"]
df
```

```python
ingreso_total = df["Ingreso"].sum()
ingreso_total
```

```python
df.groupby("Producto")["Cantidad"].sum().sort_values(ascending=False)
```

```python
df.groupby("Producto")["Ingreso"].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

ingresos_producto = df.groupby("Producto")["Ingreso"].sum()

plt.bar(ingresos_producto.index, ingresos_producto.values)
plt.title("Ingreso por producto")
plt.xlabel("Producto")
plt.ylabel("Ingreso")
plt.show()
```

## Resumen de ventas
-El producto con mayor ingreso fue zapatos
-Las camisas tienen buena rotación
-Se recomienda reforzar stock de zapatos
