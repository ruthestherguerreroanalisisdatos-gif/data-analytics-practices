```python
import pandas as pd

ventas = pd.DataFrame({
    "id_Producto": [1, 2, 3, 1, 2],
    "ventas": [10, 15, 8, 20, 5]
})

ventas
```

```python
productos = pd.DataFrame({
    "id_Producto": [1, 2, 3],
    "producto": ["Chocolate", "Vainilla", "Fresa"],
    "precio": [180, 150, 160]
})

productos
```

```python
df_inner = pd.merge(ventas, productos, on="id_Producto", how="inner")
df_inner
```

```python
df_left = pd.merge(ventas, productos, on="id_Producto", how="left")
df_left
```

```python
df_right = pd.merge(ventas, productos, on="id_Producto", how="right")
df_right
```

```python
df_outer = pd.merge(ventas, productos, on="id_Producto", how="outer")
df_outer
```

```python
df_inner["ingreso"] = df_inner["ventas"] * df_inner["precio"]
df_inner
```

```python
df_inner.groupby("producto")["ingreso"].sum()
```
