```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
  r"C:\Users\lapto\Downloads\ventas_heladeria.csv",
    sep=";",
    encoding="latin1",
    engine="python",
    on_bad_lines="skip"
)

df.head()
```

```python
import pandas as pd

datos = {
    "Producto": ["Vainilla", "Chocolate", "Fresa", "Mango", "Oreo"],
    "Precio": [150, 180, 160, 170, 200],
    "Ventas": [20, 35, 25, 30, 15]
}

df = pd.DataFrame(datos)
df
```

```python
df["Ingreso"] = df["Precio"] * df["Ventas"]
df
```

```python
ingreso_total = df["Ingreso"].sum()
ingreso_total
```

```python
precio_promedio = df["Precio"].mean()
precio_promedio
```

```python
ventas_totales = df["Ventas"].sum()
ventas_totales
```

```python
ingreso_total = df["Ingreso"].sum()
ventas_totales = df["Ventas"].sum()
precio_promedio = df["Precio"].mean()
producto_estrella = df.loc[df["Ingreso"].idxmax(), "Producto"]
```

```python
plt.figure()
plt.bar(df["Producto"], df["Ingreso"])
plt.title("Ingreso por producto")
plt.ylabel("Ingreso")
plt.show()
```

```python
plt.figure()
plt.bar(df["Producto"], df["Ventas"])
plt.title("Ventas por producto")
plt.ylabel("Unidades vendidas")
plt.show()
```

```python
plt.figure()
plt.axhline(precio_promedio)
plt.title("Precio Promedio de Productos")
plt.show()
```
