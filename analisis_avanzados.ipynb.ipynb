```python
import pandas as pd

df = pd.read_csv(r"C:\Users\lapto\Downloads\ventas_productos.csv")
df
```

```python
ventas_por_producto = df.groupby("Producto")["Ventas"].sum()
ventas_por_producto
```

```python
ingreso_por_producto = df.groupby("Producto")["Ingreso"].sum()
ingreso_por_producto
```

```python
resumen = df.groupby("Producto").agg({
    "Ventas": "sum",
    "Ingreso": "sum",
    "Precio": "mean"
})

resumen
```

```python
pivot = pd.pivot_table(
    df,
    values="Ingreso",
    index="Producto",
    aggfunc="sum"
)

pivot
```

```python
pivot.sort_values(by="Ingreso", ascending=False)
```
